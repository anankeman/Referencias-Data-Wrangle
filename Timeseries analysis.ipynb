```python
import pandas as pd
import numpy as np
import sklearn
import pandas_datareader.data as web
import matplotlib as plt
%matplotlib inline
```

```python
gs = web.DataReader("F", data_source='iex', start='2016-01-01',
                    end='2017-01-01')
```

```python
gs.open.plot(kind="line") 
```

```python
gs.head()
```

```python
gs['cambiopen']=gs.open.pct_change()
gs['cambiovol']=gs.volume.pct_change()
gs2= gs.copy()
gs2=gs2.dropna()
```

```python
gs2[['cambiopen','cambiovol']].plot(kind='line')
```

```python
gs2['cambiopen'].plot(kind='line')
```

```python
gs2.cambiopen.corr(gs2.cambiovol)
```
